Q1 Get the current day, month, year, hour, minute and timestamp from datetime module

```python
from datetime import datetime

now = datetime.now()
# Get the current day
current_day = now.day
print(f"Today is: {current_day}")

# Get the current month
current_month = now.month
print(f"Current month is: {current_month}")

# Get the current year
current_year = now.year
print(f"Current year is: {current_year}")

# Get the current hour
current_hour = now.hour
print(f"Current hour is: {current_hour}")

# Get the current minute
current_minute = now.minute
print(f"Current minute is: {current_minute}")

# Get the current second
current_second = now.second
print(f"Current second is: {current_second}")

# Get the current timestamp
current_timestamp = now.timestamp()
print(f"Current timestamp is: {current_timestamp}")


print(f'Today is: {current_day}/{current_month}/{current_year}, {current_hour}:{current_minute}')
```

Q2 Format the current date using this format: ("%m/%d/%Y, %H:%M:%S")

```python
# Format MONTH/DAY/YEAR, HOUR:MINUTE:SECOND
t = now.strftime("%m/%d/%Y, %H:%M:%S")
print(f"time: {t}")

# Format DAY/MONTH/YEAR, HOUR:MINUTE:SECOND
time_one = now.strftime("%d/%m/%Y, %H:%M:%S")
print(f"time one: {time_one}")

# Format YEAR/MONTH/DAY HOUR:MINUTE:SECOND, 
time_two = now.strftime("%Y/%m/%d, %H:%M:%S")
print(f"time two: {time_two}")

time_three = now.strftime("%H:%M:%S, %Y/%m/%d")
print(f"time three: {time_three}")

time_four = now.strftime("%H:%M:%S, %m/%d/%Y")
print(f"time four: {time_four}")

time_five = now.strftime("%H:%M:%S, %d/%m/%Y")
print(f"time five: {time_five}")
```

Q3 Today is 5 December, 2019. Change this time string to time

Q4 Calculate the time difference between now and new year.

```python
# Today's date definition
now = datetime(year = now.year, month = now.month, day = now.day, hour = now.hour, minute = now.minute, second = now.second)

# Next year to come (i.e 2026)
new_year = datetime(year = now.year + 1, month = 1, day = 1, hour=0, minute=0, second=0)

# Now minus new year
diff_Now_NewYear = now - new_year

# New year to come (i.e 2026) minus now
diff_Now_NewYear1 = new_year - now

print(f'Time difference between now and new year: {diff_Now_NewYear}\nTime difference between new year and now: {diff_Now_NewYear1}')
```

Q5 Calculate the time difference between 1 January 1970 and now

```python
# Today's date definition
now = datetime(year = now.year, month = now.month, day = now.day)
# 1 January 197
Jan_1970 = datetime(year = 1970, month = 1, day = 1)

# Now minus new year
diff_Now_1970 = now - Jan_1970

# New year to come (i.e 2026) minus now
diff_1970_now = Jan_1970 - now

print(f'Time difference between now and 1970: {diff_Now_1970}\nTime difference between 1970 and now: {diff_1970_now}')
```

Q6 Think, what can you use the datetime module for? Examples:
- Time series analysis
- To get a timestamp of any activities in an application
- Adding posts on a blog

```python
#1 Time series analysis 
#  to do that let us imports two modules that will help generate data over a time and visualize it
# Pandas will be used to generate the data
# Numpy will be used to generate random data

import pandas as pd
import numpy as np

# Generate a date range for 2 years from January 1, 2024 to December 31, 2025
date_range = pd.date_range(start='2024-01-01', end='2025-12-31', freq='D')

# Create a DataFrame with the date range
df = pd.DataFrame(date_range, columns=['Date'])

# Add a column with random data to simulate time series data
df['Value'] = np.random.randn(len(df))

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)
```

we can print part of the data, example the first to row

```python
df.head(50)
```

The statistical tools

```python
df.describe()
```

Data visualization

```python
# Perform basic time series analysis
# Calculate the rolling mean with a window of 7 days
df['RollingMean'] = df['Value'].rolling(window=7).mean()

# Plot the time series data and the rolling mean
df.plot(figsize=(12, 6), title='Time Series Analysis')
```
